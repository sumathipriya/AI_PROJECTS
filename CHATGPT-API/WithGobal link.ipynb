```python
import google.generativeai as genai
import gradio as gr

# Configure your Gemini API key
genai.configure(api_key="AIzaSyBpojGAbx9Gh6pMWN-VyMo4gbuUJPUpLrQ")

# Set the model
model = genai.GenerativeModel(model_name="models/gemini-1.5-pro-latest")

# Start a new chat session
chat = model.start_chat(history=[])

# System prompt (Gemini doesn't have a "system" role like OpenAI, but you can simulate it)
system_instruction = "You are a financial expert that specializes in real estate investment and negotiation."
chat.send_message(system_instruction)

# Define the chat function for Gradio
def CustomGeminiChat(user_input):
    response = chat.send_message(user_input)
    return response.text

# Gradio interface
demo = gr.Interface(fn=CustomGeminiChat,
                    inputs="text",
                    outputs="text",
                    title="My Saving Assistant (Gemini)")

demo.launch(share=True)
```
